```python
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#create a column age group
labels=['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi weekly':14,
    'Annually':365,
    'Every 3 months':90
    
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head(10)                                                              
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
from sqlalchemy import create_engine
import urllib

# 1. Define the connection details exactly from your screen
server_name = "YEZMA"
database_name = "customer_behavior"

# 2. Build the connection string for Windows Authentication
params = urllib.parse.quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server_name};"
    f"DATABASE={database_name};"
    f"Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# 3. Push your clean DataFrame into your SQL Server database
table_name = "customer"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Success! Data loaded into SQL Server table '{table_name}' inside '{database_name}'.")
```
